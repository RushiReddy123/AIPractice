```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
data=pd.read_csv("practice_dataset.csv")
data.head()
```

```python
data.tail()
```

```python
data["Age"].describe()
```

```python
print(data["Age"].median())
print(data["Age"].mode())
print(data["Age"].std())
print(data["Age"].var())
```

```python
range=max(data["Age"])-min(data["Age"])
print(range)
```

```python
data['Salary'].hist(bins=20,edgecolor='green')
plt.title('Histogram of salary')
plt.xlabel('Salary')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(8,5))
data['Department'].hist(bins=10,edgecolor='green')
plt.title('histogram of department ')
plt.xlabel('Departments')
plt.ylabel('frequency')
plt.show()
```

```python
deptartmentcount=data['Department'].value_counts()
print(deptartmentcount)
```

```python
plt.figure(figsize=(8,5))
plt.pie(deptartmentcount,labels=deptartmentcount.index,autopct='%1.1f%%',startangle=90)
plt.title('Pie chart ')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(data["Age"], bins=30, density=True, color='skyblue', edgecolor='black')
plt.title('Central Limit Theorem: Distribution of Sample Means')
plt.xlabel('Sample Mean')
plt.ylabel('Density')
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```
